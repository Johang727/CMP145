### 1.2 Count the Names

Now, use a Python dictionary to count how many times each name occurs in the list `students`.

````python
# TODO: use a dictionary to count how many times each name occurs in students
nameCounts = {}

with open('students.csv', 'r') as students:
    students_csv = csv.reader(students)
    for line in students_csv:
        nameCounts[line[0]] = nameCounts.get(line[0], 0) + 1

print(nameCounts)
````

### 1.3 Rank the Names

Now that you have a dictionary with the frequency of each name, get a list of the `(name, count)` tuples of the dictionary, and sort the list in descending order by count. Then, print out the ten most popular names in the list.

````python
# TODO: get a list of (name, count) tuples from your name frequency dictionary

student_list = []

for item in nameCounts.items():
    student_list.append(item)

def getFreq(t):
    return t[1]


student_list.sort(key = getFreq, reverse=True)


# TODO: print the top ten most popular names and their frequencies

for i in range(10): 
    print(student_list[i])
````
