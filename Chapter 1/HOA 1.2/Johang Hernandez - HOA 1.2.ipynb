### 2.3 Rectanglesville!

Now copy the code of `drawSquare()` again in the cell below, and modify it so it draws a rectangle whose width is twice the sideLength. Then, execute the cell that follows to test your modifications. 

````python
# copy drawSquare() again, modify to draw a rectangle
# copy the drawSquare() function here
def drawSquare(myTurtle, sideLength):
    for i in range(2):
        myTurtle.forward(sideLength)
        myTurtle.right(90)
        myTurtle.forward(sideLength*2)
        myTurtle.right(90)
````

### 2.4 Better Rectangles!

Now, write a *generalized* rectangle function, `drawRectangle()`, that takes three parameters: the turtle, the width, and the height. The function has been started for you in the cell below.

````python
# write drawRectangle() here
def drawRectangle(myTurtle, width, height):
    for i in range(2):
        myTurtle.forward(height)
        myTurtle.right(90)
        myTurtle.forward(width)
        myTurtle.right(90)
````
